```python
spotify['track_id'].nunique()
```

```python
#Top 10 Artist 
Top_Artist = spotify['track_artist'].value_counts()[:10]
Top_Artist
```

```python
plt.figure(figsize=(22,6))
plt.bar(Top_Artist.index,Top_Artist.values,color = ['yellow','blue','purple','orange'])
plt.title("Top Track Artist")
plt.xlabel("Artist Name")
plt.ylabel("Value")
plt.show()
```

Martin Garrix has the maximum number of Songs in the dataset followed by Queenn,The Chainsmokers,David Guetta etc

```python
# Top 10 most played Track
top_track_name = spotify['track_name'].value_counts()[:10]
top_track_name
```

```python
plt.figure(figsize = (25,5))
plt.bar(top_track_name.index,top_track_name.values , color = ['yellow','brown','green'])
plt.title("Top Track Name")
plt.show()
```

Poison is most played Track.

```python
plt.figure(figsize = (27,6))
g = df['track_album_name'].value_counts()[:10]
plt.bar(g.index , g.values , color = ['lightgreen','green'])
plt.title("Top Album Name")
plt.show()
```

The title of the best album is "Greatest Hits."

```python
year = pd.to_datetime(spotify['track_album_release_date']).dt.year
```

```python
# Top releases by year
plt.figure(figsize=(10,7))

top_year = year.value_counts()[:10]
plt.pie(top_year.values,labels = top_year.index , autopct = "%.2f%%")
plt.title("Top Release Year ")
plt.show()
```

2019 releases the most tracks, followed by 2018

```python
Month = pd.to_datetime(df['track_album_release_date']).dt.month
month_release = Month.value_counts().sort_index()
month_release
```

```python
plt.figure(figsize = (20,5))
plt.bar(month_release.index,month_release.values,color = ['red','blue','green','orange','magenta'])
plt.xlabel("Month")
plt.ylabel("Track album release")
plt.show()
```

First month of the year i.e. January has the maximum release followed by November,October and September.

```python
genre = spotify['playlist_genre'].value_counts()
plt.pie(genre,labels = genre.index , autopct = "%.2f%%" , colors = ['yellow', 'lightblue', 'lightgreen', 'orange', 'grey','blue',] )
plt.title("Track popularity by genre")
plt.show()
genre 
```

18.4% of tracks in our dataset is of 'edm' genre.
17.5% of tracks are from 'rap' genre.

```python
sub_genre = spotify['playlist_subgenre'].value_counts()[:5]
plt.pie(sub_genre,labels = sub_genre.index , autopct = "%.2f%%" , colors = ['yellow', 'lightblue', 'lightgreen', 'orange', 'grey','blue',] )
plt.title("Track popularity by Sub Genre")
plt.show()

for i,count in sub_genre.items():
    print(i,":" ,count)
```

progressive electro house subgenre has the maximum songs followed by southern hip hop, indie poptimism and so on.

```python
numeric_columns = spotify.select_dtypes(include=['float64', 'int64'])
plt.figure(figsize=(10, 8))  # Set the size of the heatmap
sns.heatmap(numeric_columns.corr(), annot= True)
plt.title('Correlation Heatmap')
plt.show()
```

Danceability describes how suitable a track is for dancing based on a combination of musical elements including tempo, rhythm stability, beat strength, and overall regularity. A value of 0.0 is least danceable and 1.0 is most danceable.

```python
most_dancebility = df[['track_name','track_artist','danceability']].sort_values(by = 'danceability',ascending = True)[:5]
most_dancebility
```

```python
plt.figure(figsize = (25.,5))
plt.bar(most_dancebility.track_artist , most_dancebility.danceability , color = ['black','grey'])
plt.show()
```

```python
least_dancebility= df[['track_name','track_artist','danceability']].sort_values(by = 'danceability',ascending = False)[:5]
least_dancebility
```

```python
plt.figure(figsize = (25.,5))
plt.bar(least_dancebility.track_artist , least_dancebility.danceability , color = ['orange','brown'])
plt.show()
```

 Hi, How're You Doin'? value  is least danceable and If Only I Could (feat. Steve Lucas) - Liem Remix is most danceable.

Duration of song in milliseconds

```python
duration_albums = spotify.nlargest(5,'duration_ms')

top_duration_albums = duration_albums[['track_name','duration_ms']]
```

```python
plt.figure(figsize = (25,5))
plt.bar(top_duration_albums['track_name'] , top_duration_albums['duration_ms'] , color = ['yellow','lightblue', 'lightgreen','orange'] )
plt.show()
```

The longest duration is found in the remix of the song titled "47."

```python
duration_albums = spotify.nsmallest(5,'duration_ms')

least_duration_albums = duration_albums[['track_name','duration_ms']]
plt.figure(figsize = (25,5))
plt.bar(least_duration_albums['track_name'] , least_duration_albums['duration_ms'] , color = ['lightblue','blue'] )
plt.show()
```

The least duration is found in the remix of the song titled "Hi,How're You Doin'?"
